```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests

import sys
import random
import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://m.imdb.com/chart/top/'

headers = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}

res = requests.get(url, headers=headers)
html_page = res.content

soup = BeautifulSoup(html_page, 'html.parser')
movies = soup.find_all('li', {'class' : 'ipc-metadata-list-summary-item sc-3f724978-0 enKyEL cli-parent'})
print(len(movies))
```

```python
def progressBar(count_value, total, start_time, end_time, suffix=''):
    bar_length = 70
    start = start_time
    filled_up_Length = int(round(bar_length* count_value / float(total)))
    percentage = round(100.0 * count_value/float(total),1)
    bar = '=' * filled_up_Length + '-' * (bar_length - filled_up_Length)
    
    elapsed_time = (end_time - start_time).total_seconds()
    
    sys.stdout.write('[%s] %s%s completed. Time: %s secs ...%s\r' %(bar, percentage, '%',elapsed_time, suffix))
    sys.stdout.flush()
```

```python
top_shows = pd.DataFrame(columns=['Rank', 'Title', 'Released', 'Rating', 'Genres', 'Description'])
start_time = datetime.datetime.now()
for i in range(len(movies)):
    rank = movies[i].h3.text.split('.')[0].strip()
    title = movies[i].h3.text.split('.')[1].strip()
    released = movies[i].find('div', class_ = 'sc-43986a27-7 dBkaPT cli-title-metadata').text[:4]
    #duration = movies[i].find('div', class_ = 'sc-43986a27-7 dBkaPT cli-title-metadata').text[4:10]
    #certification = movies[i].find('div', class_ = 'sc-43986a27-7 dBkaPT cli-title-metadata').text[10:])
    rating = movies[i].find('div', class_ = 'sc-9ab53865-0 bIaPYM sc-43986a27-2 bvCMEK cli-ratings-container').text[:3]
    movie_link = 'https://m.imdb.com/'+movies[i].a.get('href')
    movie_res = requests.get(movie_link, headers=headers)
    movie_info = BeautifulSoup(movie_res.content, 'html.parser')
    movie_genre = movie_info.find('div', class_ = 'ipc-chip-list__scroller').find_all('span', class_="ipc-chip__text")
    genres = []
    for j in range(len(movie_genre)):
        genres.append(movie_genre[j].get_text())
    
    try:
        desc = movie_info.find('div', class_ = 'sc-e226b0e3-10 hbBxmX').p.span.text
    except AttributeError:
        desc = movie_info.find('div', class_ = 'sc-e226b0e3-10 kRcXhr').span.text
    
    #desc = movie_info.find('div', class_ = 'sc-e226b0e3-10 hbBxmX').p.span.text
        
    top_shows = top_shows.append({'Rank': rank,
                      'Title': title,
                      'Released' : released,
                      'Rating': rating,
                      'Genres': genres,
                      'Description': desc}, ignore_index=True)
    
    end_time = datetime.datetime.now()
    
    progressBar(i+1,len(movies), start_time, end_time)
    #print('Movie: {} inserted'.format(i+1))
```
